```python
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader
```

```python
## 데이터셍을 만들 이미지 정보 설정
img_root    = '../Data/animals'
train_dir   = f'{img_root}/train'
test_dir    = f'{img_root}/test'
valid_dir = f'{img_root}/val'
```

```python
## 학습용/검증요/테스트용 이미지 dataset 생성
from torchvision import transforms

trainDS = ImageFolder(root=train_dir, transform=transforms.ToTensor)
validDS = ImageFolder(root=valid_dir)
testDS = ImageFolder(root=test_dir)


print(f'trinaDS calsses : {len(trainDS)}개, {trainDS.classes}, {trainDS.class_to_idx}')
print(f'validDS calsses : {len(validDS)}개, {validDS.classes}, {validDS.class_to_idx}')
print(f'testDS calsses : {len(testDS)}개, {testDS.classes}, {testDS.class_to_idx}')
```

```python
## dataLodoader

trianDL = DataLoader(trainDS,batch_size=1)
for img, label in trianDL :
    print(label)
    
```
