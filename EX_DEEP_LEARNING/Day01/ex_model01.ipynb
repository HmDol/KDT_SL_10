### [ANN 모델 구성]
- 기본구성 : 입력층 - 은닉층 - 출력층
- Linear Layer 클래스
    - 퍼셉트론의 가중치 합계 기능처리 클래스
    - 입력값으로 피쳐 수
    - 출력값으로 출력 수 ==> 층의 퍼셉트론 수
- nn.Module 클래스
    - 커스텀 모델 구현을 위한 부모 클래스

[1] 모듈 로딩 <hr>

```python
import torch                        ## 텐서 및 수치함수들
import torch.nn as nn               ## 인공신경망 모듈
import torch.nn.functional as F     ## 인공신경망 관련 함수들 모듈
```

[2] 모델 설계<hr>

```python
## -----------------------------------------------------------------
## 주제 : 거리, 신호, 날씨에 따른 도착시간 예측 모델
## 피쳐 : 거리, 신호, 날씨
## 타겟 : 도착시간
## 종류 : 지도학습 + 회귀
## -----------------------------------------------------------------
##          입력     퍼셉트론    출력        활성함수
## 입력층     3         10       10         ReLu
##           ____________________|
##           ↓
## 은닉층     10        5        5          ReLu
##           ___________________|
##           ↓
## 출력층     5         1        1          x(회귀라서)
## -----------------------------------------------------------------
## 클래스이름 : annModel
## 부모클래스 : nn.Module
## 오버라이딩 : __init__()  : 모델 층 구성
##            forward()   : 순방향 학습 
## ----------------------------------------------------------------
class annModel(nn.Module) :
    ## 모델 층 구성 및 초기화 메서드
    def __init__(self) :
        super().__init__()
        self.in_layer = nn.Linear(3, 10) ## 입력 3개 출력 10개
        self.hd_layer = nn.Linear(10, 5) ## 입력 10개 출력 5개
        self.ot_layer = nn.Linear(5, 1) ## 

##순방향 진행 메서드
def forward(self, x) :
    ## 입력층 => 가중치합계 => AF
    out = self.in_layer(x)
    out = F.relu(out)
    
    ## 입력층 => 가중치합계 => AF
    out = self.hd_layer(out)
    out = F.relu(out)
    
    ## 출력층 => 가중치합계 Becasue .회귀
    out = self.ot_layer(out)
    
    return out
    
```
