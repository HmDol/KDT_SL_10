#### [Perceptron 이해]
- 입력 데이터와 가중치 곱셈의 합계 + 절편 결과 반환
- 모든 입력 데이터를 가중치와 곱셈 연산 수행
- 연산 결과를 활성화함수(AF)를 거쳐서 최종 결과값 추출
- 뉴런의 기본 동작을 코드로 구현한 것

[1] 모듈 로딩 및 데이터 준비

```python
## [1-1] 모듈로딩
import torch                        # 텐서 및 수치함수들 관련 모듈
import torch.nn as nn               # 인공신경망 관련 모듈
import torch.nn.functional as F     # 인공신경망 관련 함수들 모듈(AF, LF ,MF....)
```

```python
## [1-2] 데이터 준비
## 랜덤 고정
torch.manual_seed(2)

## 입력 데이터 텐서 생성
xTS = torch.randint(1, 10, (3, 1)) ## 1 ~ 9까지 3행 1열 랜덤 텐서
xTS = xTS.type(torch.float32) ## int텐서 -> float 텐서로 형변환 : tensor.type(변환)

print(xTS.shape, xTS.ndim, xTS.dtype)
print(xTS)
```

```python
## 정답 텐서 생성
yTS = torch.randint(1, 10, (3, 1))
yTS = yTS.type(torch.float32)

print(yTS.shape, yTS.ndim, yTS.dtype)
print(yTS)
```

[2] 가중치와 절편 텐서 생성

```python
## 가중치와 절편 텐서 값을 0으로 초기화
## 퍼셉트론 1개, 입력 피쳐 1개, 샘플 3개
## 가중치와 절편 텐서 역전파로 업데이트 되어야 함 => requires_grad=True
W = torch.rand(1, requires_grad=True)
b = torch.rand(1, requires_grad=True)

print(W, b)
```

[3] 연산 => 입력 데이터 * 가중치 + 절편

```python
output = xTS * W + b
output
```

[4] 정답과 연산결과 값의 차이 계산 => 손실함수

```python
# 정답
print("정답\n", yTS)

## 학습 결과 값
print("결과값\n", output)

## 정답 - 학습결과값
value = yTS-output
print("차이값\n", value)

#차이값 제곱 후 평균 : Mean Square Error MSE 평균제곱 오차
value = (yTS - output)**2
value = torch.mean(value)
print("MSE\n", value)
```
