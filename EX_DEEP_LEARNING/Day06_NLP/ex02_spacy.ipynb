```python
import spacy

nlp = spacy.load("ko_core_news_lg")

import ko_core_news_lg

nlp = ko_core_news_lg.load()
doc = nlp("이것은 문장입니다.")

print([(w.text, w.pos_) for w in doc])
```

```python
## 형태소 분석 진행
text = "일만 하고 놀지 않으면 바보가 된다. 김힘찬, 임정현"
ret = nlp(text)
print(f"ret -> {type(ret)}")
```

```python
for token in ret :
    print(f'text 속성 :{token.text:<10}품사 : {token.pos_:<8}', end='')
    print(f'상세 :{token.tag_:<10} 어근 : {token.lemma_:<5}')
        
print("\n명사만 추출")
for token in ret :
    if token.pos_ in ['NOUN', 'PROPN'] :
        print(f'text 속성 :{token.text:<10}품사 : {token.pos_:<8}', end='')
        print(f'상세 :{token.tag_:<10} 어근 : {token.lemma_:<5}')
```
