[2] nn.Sequential 클래스 <hr>

```python
## =============================================
## [2-1] 기본 형태
## =============================================
## 모델 인스턴스 생성
model = nn.Sequential( 
    nn.Linear(3, 10),
    nn.ReLU(),
    nn.Linear(10, 5),
    nn.ReLU(),
    nn.Linear(5, 1)
    )
```

```python
## 모델 구성 요소
print("모델 층 구조", model)

print("\n\n===parameters===")
for name, param in model.named_parameters():
    print(f"모델 층별 파라미터 {name}=>", param.shape )
    

print("\n\n===child-module===")
for name, child in model.named_children():
    print(f"child {name}=>", child )


print("\n\n===module===")
for module_ in model.modules():
    print("module_=>", module_ )
```

```python
## 순방향
model( torch.tensor([[1,2,3]], dtype=torch.float32) )
```

```python
## 모델 요약 정보
summary(model, input_size=(1, 3))
```

```python
## ===============================================
## [2-2] OrderedDict타입으로 층 구성 : 층에 이름 부여
## ===============================================
## 모델 인스턴스 생성
model = nn.Sequential( 
    OrderedDict([
        ('h1',      nn.Linear(3, 10)),
        ('relu1',   nn.ReLU()),
        ('h2',      nn.Linear(10, 5)),
        ('relu2',   nn.ReLU()),
        ('out',     nn.Linear(5, 1)),
    ])
)
```

```python
## 모델 구성 요소
print("모델 층 구조", model)
```

```python
summary(model, input_size=(1, 3))
```
