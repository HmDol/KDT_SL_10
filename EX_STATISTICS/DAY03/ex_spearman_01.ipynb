#### [Spearman Rank Coefficient]
- 연속형 데이터, 선형 또는 비선형 단조 강도 측정
- 단조 : 전체적으로 데이터의 증가 또는 감소
- 정규분포 x, 이상치 여향 약, 비선형도 가능

```python
## ===============================================================
## 모듈 로딩
## ===============================================================
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr
import koreanize_matplotlib
```

```python
## 재현성
np.random.seed(0)
```

```python
## ==================================================================
## 선형 단조 증가 (linear, monotonic)
## ==================================================================
x = np.linspace(-2, 2, 80)
y_linear = 2 * x + 1 + np.random.normal(scale = 0.5, size= len(x))
# print(x)
# print(y_linear)
## 2 변수에 대한 선형 관계성(pearson), 단조 관계성 (spearman) 측정
pearson_linear, _ = pearsonr(x, y_linear)
spearman_linear, _ = spearmanr(x, y_linear)

## 시각화
plt.figure()
plt.scatter(x, y_linear)
plt.title(f'선형 단조 증가\nPearson={pearson_linear : .3f}, Spearman={spearman_linear:.3f}')
plt.xlabel('X')
plt.ylabel('Y(linear)')
plt.grid(True)
plt.show()

## 수치만 따로 출력
print("상관계수 비교 - 선형 단조 증가")
print(f'* Pearson : {pearson_linear:.4f}')
print(f'* spearman : {spearman_linear:.4f}')
```

```python
## ==================================================================
## 비선형ㄴ 단조 증가(nonlinear, monotonic)
##  y = x^5 + 작은 노이즈 -> 방향은 계속 증가, 모양은 확 휘어짐
## ==================================================================
x = np.linspace(-2, 2, 80)
y_mono = x**5 + np.random.normal(scale = 0.5, size= len(x))
# print(x)
# print(y_mono)
## 2 변수에 대한 선형 관계성(pearson), 단조 관계성 (spearman) 측정
pearson_mono, _ = pearsonr(x, y_mono)
spearman_mono, _ = spearmanr(x, y_mono)

## 시각화
plt.figure()
plt.scatter(x, y_mono)
plt.title(f'비선형 단조 증가\nPearson={pearson_mono : .3f}, Spearman={spearman_mono:.3f}')
plt.xlabel('X')
plt.ylabel('Y(nonlinear monotonic)')
plt.grid(True)
plt.show()

## 수치만 따로 출력
print("상관계수 비교 - 선형 단조 증가")
print(f'* Pearson : {pearson_mono:.4f}')
print(f'* spearman : {spearman_mono:.4f}')
```

```python
## ==================================================================
## 비단조 관계 (non-monotonic)
##  y = sin(x) -> 올라갔다 내려가는 형태
## ==================================================================
x = np.linspace(0, 10, 80)
y_non = np.sin(x) + np.random.normal(scale = 0.1, size= len(x))
# print(x)
# print(y_mono)
## 2 변수에 대한 선형 관계성(pearson), 단조 관계성 (spearman) 측정
pearson_non, _ = pearsonr(x, y_non)
spearman_non, _ = spearmanr(x, y_non)

## 시각화
plt.figure()
plt.scatter(x, y_non)
plt.title(f'비단조 형태\nPearson={pearson_non : .3f}, Spearman={spearman_non:.3f}')
plt.xlabel('X')
plt.ylabel('Y(non-monotonic)')
plt.grid(True)
plt.show()

## 수치만 따로 출력
print("상관계수 비교 - 선형 단조 증가")
print(f'* Pearson : {pearson_non:.4f}')
print(f'* spearman : {spearman_non:.4f}')
```
