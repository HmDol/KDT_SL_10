#### [Spearman Rank Coefficient]
- 광고비 <-> 웹사이트 방문자 수 : 비선형 단조 증가
- 현실에서는 광고비를 2배 쓰면 방문자가 반드시 2배  증가하지 않음
    - 증가하긴 하는데 증가폭은 점점 줄어드는 비선현 단조 증가 관계가 대부분

```python
## ===============================================================
## 모듈 로딩
## ===============================================================
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr
import koreanize_matplotlib
```

```python
## 광고비 (100만원 단위)
ad_cost = np.array([100, 200, 300, 400, 500, 600, 700, 800, 900, 1000])

## 방문자 수 : 앞 9개는 거의 직선, 마지막 1개만 '대박 캠페인'
visitor = np.array([1000, 2000, 3000, 4000, 5000,
                    6000, 7000, 8000, 9000, 50000])
```

```python
# 관계성 계산
pearson, _ = pearsonr(ad_cost, visitor)
spearman, _ = spearmanr(ad_cost, visitor)

plt.scatter(ad_cost, visitor)
plt.title(f'광고비 ↔ 방문자수(대박 캠페인 1개 포함)\nPearson={pearson : .3f}, Spearman={spearman:.3f}')
plt.xlabel('광고비(만원)')
plt.ylabel('방문자수(명)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd

dataDF = pd.DataFrame({'ad_cost' : ad_cost, 'visitor' : visitor})
display(dataDF.corr(method='spearman'), dataDF.corr() ) ## 기본 : pearson
```
