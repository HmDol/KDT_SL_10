### [ 기술통계/요약통계 - 데이터의 특성을 수치로 설명하는 것 ]

[1] 모듈 로딩 및 데이터 준비 <hr>

```python
## 모듈 로딩
import pandas as pd
import numpy as np
import wquantiles as wq
```

```python
## 임의의 데이터
a_data  = [70, 72, 75, 75, 76, 80, 82, 85, 88, 90]
a_sr = pd.Series(a_data)
```

[2] 데이터 확인 <hr>

```python
## 기술통계값 확인
a_sr.describe(percentiles=[0.3, 0.6, 0.9])
```

```python
## =========================================================================
## 평균/분산/표준편차 : pandas와 numpy의 분산/표준편차 계산식의 차이로 값 다룸
## =========================================================================

print('평균  중앙값       분산')
## => pandas
mean_pd = a_sr.mean()
var_pd = a_sr.var(ddof=0)
std_pd = a_sr.std(ddof=0)
print(mean_pd, var_pd, std_pd)

## => numpy
mean_np = np.mean(a_sr)
var_np = np.var(a_sr)
std_np = np.std(a_sr)
print(mean_np, var_np, std_np)

```
