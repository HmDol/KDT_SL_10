```python
## 정답!
# ======================================================================
# Q18. (실습 — 전처리 기초)  
# ======================================================================
# `titanic.csv`를 불러와 다음을 수행하시오.

# 1) Age 결측치를 중앙값으로 대체하고, 결과를 `df_age_filled`라는 DataFrame에 저장  
# 2) Fare 항목의 히스토그램을 그리시오  
# 3) Fare에서 이상치(상위 1%) 값의 개수를 계산하여 `q18_outlier_count` 변수에 저장  

# 아래 코드 셀에 작성하세요.
```

```python
## 데이터 불러오기
data_path = './titanic.csv'
df = pd.read_csv(data_path)
# df.info()
# df.head()
# 1) Age 결측치를 중앙값으로 대체하고, 결과를 `df_age_filled`라는 DataFrame에 저장  
tempSR = df['Age'].fillna(df['Age'].median())
df_age_filled = df.copy()
df_age_filled['Age'] = tempSR
df_age_filled.info()

# 2) Fare 항목의 히스토그램을 그리시오
plt.hist(df['Fare'], bins=30, color='blue', alpha=0.7)
plt.title('Fare Histogram')
plt.show()

# 3) Fare에서 이상치(상위 1%) 값의 개수를 계산하여 `q18_outlier_count` 변수에 저장
q18_outlier_count =len(df[df['Fare'] > df['Fare'].quantile(0.99)])
print('이상치(상위 1%) 개수 :' , q18_outlier_count)
```

```python
## 정답!
# ======================================================================
# Q19. (실습 — 분포 시각화)  
# ======================================================================
# `titanic.csv` 데이터에서 다음을 수행하시오.
# 1) Age와 Fare 각각에 대해 히스토그램과 Boxplot을 그리시오.  

# 2) 두 변수의 분포 특징(대칭/비대칭, 이상치 존재 여부 등)을 아래 마크다운 셀에 간단히 서술하시오.
```

```python
import koreanize_matplotlib
# 1) Age와 Fare 각각에 대해 히스토그램과 Boxplot을 그리시오.  
for col in ['Age', 'Fare'] :
    plt.figure(figsize=(12, 5))
    
    ## 히스토그램
    plt.subplot(1, 2, 1)
    plt.hist(df_age_filled[col], bins=30, color='blue')
    plt.title(f'{col} Histogram \n첨도 : {stats.kurtosis(df_age_filled[col]):.2f} / 왜도 : {stats.skew(df_age_filled[col]):.2f}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    

    ## 박스플롯
    plt.subplot(1, 2, 2)
    plt.boxplot(df_age_filled[col])
    plt.title(f'{col} Boxplot')
    plt.ylabel(col)

    plt.tight_layout()
    plt.show()
```
