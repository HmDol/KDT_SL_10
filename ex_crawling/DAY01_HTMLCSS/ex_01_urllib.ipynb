[2-1] urlretrieve 경우 ==> url의 데이터를 파일로 저장

```python
## ========================================================
## [2-1] urlopen 경우 => url 테이터를 파일로 저장
## ========================================================
img_url = r'https://search.pstatic.net/common/?src=http%3A%2F%2Fblogfiles.naver.net%2FMjAyNDExMTVfMTAz%2FMDAxNzMxNjcyNzMzMTQz.Y3Jf1YJi5Omke-_PnXZuEDs-cshcoP9vnlyRLHmEkuYg.u8AZQ-0kyIVMbijBozNTRbKwXB9bbDgQdrjpD2BPEJYg.JPEG%2F20241115-210801-012.jpg&type=sc960_832'
save_file = '../Data/siba1.jpg'

f_path , msg = urlretrieve(img_url,save_file)

if os.path.exists(save_file) :
    print(f'f_path =>{f_path}')
    print(f'msg => {msg}')
else :
    print('url 체크 바랍니다. 저장 실패!')
```

```python
## ========================================================
## [2-2] urlopen 경우 => url 테이터를 저장
## ========================================================

web_url = r'https://www.naver.com/'
# imgReq = urlopen(img_url)
imgReq = urlopen(web_url)

print(f'imgReq = > {imgReq}')
print(f'i|mgReq.headers = > {imgReq.headers}')
print(f'메서드 : {imgReq.read()}')
```

```python
## 이미지 로우 데이터 읽어서 저장
resObj =imgReg = urlopen(img_url)
## 바이너리 데이터 추출
raw_data = resObj.read()
save_file = '../Data/siba2.jpg'

## 바이너리 모드로 파일 open * write
with open(save_file, mode = 'wb') as f:
    f.write(raw_data)
    print('저장되었습니다.')
```

```python
## =============================================================
## 텍스트 로우 데이터 읽어서 저장
## =============================================================

# url 과 저장 경로 지정
data_url = 'http://api.aoikujira.com/ip/ini'

## 다운로 드 / 데이터 읽어 들이기
res = urlopen(data_url)
data = res.read()

## 바이트 데이터 읽어오기
print('tpye(data)', type(data))

## 바이너리 => 인코딩해서 str로 저장
text = data.decode('utf-8')
print('tpye(text)', type(text))
print(text)
```

[3] url 관련 예외처리<hr>

```python
## 모듈 로딩
from urllib.request import urlopen
from urllib.error import HTTPError, URLError

## url 데이터 로딩
try : 
    html=urlopen('http://www.pythonscraping.com/pages/error.html')
except HTTPError as e :
    print(f'error msg : {e}')
except URLError as u :
    print(f'error msg : {u}')
else :
    print("ok", html.read(), sep = '\n\n')
finally :
    print("-----end-----")
```
