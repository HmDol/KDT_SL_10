```python
from urllib.request import urlopen          ## 웹에서 url기반으로 다운로드
from urllib.request import Request          ## 요청 객체 즉, requestMessage 생성
from bs4 import BeautifulSoup               ## 웹데이터 객체화 및 정보 추출용 모듈
```

```python
## bs 객체 생성
url_path = r'https://ko.wikisource.org/wiki/%EC%A0%80%EC%9E%90:%EC%9C%A4%EB%8F%99%EC%A3%BC?page=2&sort=title#section1'
## Request 객체에 User-Agent 헤더 추가
req = Request( url_path, 
               headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'})



html    = urlopen(req)
print(html)
bs      = BeautifulSoup(html, 'html.parser')
```

```python
## 전역 변수
arr = []
```

```python
def func(href):
    path = "https://ko.wikisource.org" + href

    req = Request(path, headers={'User-Agent': 'Mozilla/5.0'})
    html = urlopen(req).read()
    bs = BeautifulSoup(html, 'html.parser')

    ## 제목 추출
    title_node = bs.select_one('.header-section-text')
    title = title_node.get_text(strip=True) if title_node else "제목없음"
    print("제목:", title)

    ## 시 본문 추출
    poem_node = bs.find("div", class_='prp-pages-output').text
    # poem_text = poem_node.get_text("\n", strip=True) if poem_node else "본문없음"
    print("내용:")
    print(poem_node)
    print("-" * 50)

    ## 배열에 저장
    arr.append([title, poem_node])
```

```python
#mw-content-text > div.mw-content-ltr.mw-parser-output >  a
#mw-content-text > div.mw-content-ltr.mw-parser-output > ul:nth-child(14) > li:nth-child(2) > a
#mw-content-text > div.mw-content-ltr.mw-parser-output > ul:nth-child(12) > li:nth-child(3) > a

temp = bs.select('.mw-heading.mw-heading3 ~ ul a')
arr = []
for t in temp :
    print(t.attrs['href'])
    func(t.attrs['href'])
```

```python
import pandas as pd

df = pd.DataFrame(arr)
df.columns = ['제목', '본문']

df.to_csv('윤동주_시.csv')
```
