### [ 웹 스크래핑 - 정보추출 ]
- 수집된 웹 데이터에서 필요한 정보 추출하는 작업 : 스크래핑
- 스크래핑을 편리하게 진행하기 위한 패키지 : BeautifulSoup4
- binary/str 타입의 웹 데이터 ==> 객체화 (DOM)해서 프로그램 쉽게 사용할 수 있도록 함

[1] 모듈 로딩 <hr>

```python
from urllib.request import urlopen      ## 웹에서 url기반으로 다운로드
from bs4 import BeautifulSoup           ## 웹데이터 객체화 및 정보 추출용 모듈
```

[2] markup 문자열에서 정보 추출<hr>
- find()_all() 메서드
    * 조 건에 해당하는 모든 태그들 list형태로 변환
    

```python
## html 파일에서 일부분 선택 str로 저장
html_str = '''
<body>
    <a class="ca main" href="http://www.naver.com">네이버</a>    
    <a class="cb" href="http://www.google.com" target="balck">구글</a>
<h1>풀꽃</h1>
<p class="ca" id='p1'>
[경력]
2020.04 ~ 제43대 한국시인협회 회장<br>
</p>
 
<p id='p2'>
나태주                                                                    
1945년 출생으로 대표 시 「풀꽃」처럼 작고 여린 존재를 향한 시를 쓴다. 
</p>
<p id='p3'>
1963년 공주사범학교를 졸업한 뒤 초등학교에서 43년간 교직 생활을 하다가 2007년 장기초등학교 교장으로 정년퇴임했으며
 </p>
<p id='p4'>
흙의문학상, 박용래문학상, 편운문학상, 한국시인협회상, 정지용문학상, 유심작품상 등을 받았고, 
<br>

</body>
'''    
```

```python
## 마크업 str = 객체화
mbs = BeautifulSoup(html_str, 'html.parser')
```

```python
# find_all() 메서드 활용 :
h1Tag  = mbs.find_all('h1')
print(f'h1Tag => {len(h1Tag)}개, ', h1Tag)
print("=="*30)

pTag  = mbs.find_all('p')
print(f'pTag => {len(pTag)}개, ', pTag)
print("=="*30)

aTag = mbs.find_all('a')
print(f'aTag => {len(aTag)}개, ', aTag)
```

```python
# [2] find_all() 메서드 활용 : bs객체명.find_all(attr_dict, **kwargs)

## => tag의 속성으로 검
caTag = mbs.find_all(class_ = 'ca')
print(f'caTag => {len(caTag)}개, ')
for tag in caTag :
    print(tag)

```

```python
## [3] find()_all 메서드 활용 : bs객체명.find(text) -> string 변환

textTags = mbs.find_all(string='풀꽃')
print(f'textTags => {len(textTags)}개, ')
for tag in textTags :
    print(textTags)
```

```python
## =>
a2 = mbs.find(attrs={'href' : 'http://www.naver.com'})
print('a2 : ', a2)
```
