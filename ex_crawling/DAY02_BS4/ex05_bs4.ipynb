### [ 웹 스크래핑 - 정보추출 ]
- 수집된 웹 데이터에서 필요한 정보 추출하는 작업 : 스크래핑
- 스크래핑을 편리하게 진행하기 위한 패키지 : BeautifulSoup4
- binary/str 타입의 웹 데이터 ==> 객체화 (DOM)해서 프로그램 쉽게 사용할 수 있도록 함

[1] 모듈 로딩 <hr>

```python
from urllib.request import urlopen      ## 웹에서 url기반으로 다운로드
from bs4 import BeautifulSoup           ## 웹데이터 객체화 및 정보 추출용 모듈
```

[2] markup 문자열에서 정보 추출<hr>
- select_one() 메서드
    - 조건에 맞는 첫번째 tag반환
    - find()메서드와 검색 결과 동일

- select() 메서드
    - 조건에 맞는 모든 tag list로 반환
    - find_all() 메서드와 검색 결과 동일

- css 선택자
    - tag의 id 속성 : #id명
    - tag의 class 속성 : .class명
    - 자식 태그 : 부모태그 > 자식태그
    - 자손 태그 : 상위/부모태그 &nbsp;&nbsp; 하위/자속태그

    

```python
## html 파일에서 일부분 선택 str로 저장
html_str = '''
<html lang="en">
<body>
    <div id="meigen">
        <h1> 위키북스 도서 </h1>
        <ul class ='items'>
            <li>유니티 게임 이펙트 입문</li>
            <li>스위프트로 시작하는 아이폰 앱 개발 굑5ㅘ서</li>
            <li>보던 웹사이트 디자인 정석</li>

        </ul>
    </div>
</body>
</html>
'''    
```

```python
## 마크업 str = 객체화
mbs = BeautifulSoup(html_str, 'html.parser')
```

```python
## => 1개만 추출 : 타이틀 부분
h1 = mbs.select_one("div#meigen > h1").string
print(h1)


li_list = mbs.select_one('div#meigen > ul.items > li')
for li in li_list :
    print(li)
```
