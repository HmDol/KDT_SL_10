### [ 웹 스크래핑 - 정보추출 ]
- 수집된 웹 데이터에서 필요한 정보 추출하는 작업 : 스크래핑
- 스크래핑을 편리하게 진행하기 위한 패키지 : BeautifulSoup4
- binary/str 타입의 웹 데이터 ==> 객체화 (DOM)해서 프로그램 쉽게 사용할 수 있도록 함

[1] 모듈 로딩 <hr>

```python
from urllib.request import urlopen      ## 웹에서 url기반으로 다운로드
from bs4 import BeautifulSoup           ## 웹데이터 객체화 및 정보 추출용 모듈
```

[2] WEB에서 데이터 다운로드 <HR>

```python
## 다운로드 받을 웹 사이트 url
w_url = 'https://quotes.toscrape.com/'

## 다운로드
retObj = urlopen(w_url)
print(f'type(retObj) =>  {type(retObj)}')

## 로우데이터만 가져오기
bData = retObj.read() ## 한번 read하면 다음 read안됨 ==> 변수 저장 필수!!
display(bData)
```

[3] WEB 데이터에서 원하는 데이터 추출 : 스크래핑<hr>

```python
## BS객체 생성 : 마크업언어 스타일 데이터 => 사람이 이해하기 쉽게 객체화
bs = BeautifulSoup(bData, 'html.parser')
print(type(bs))
```

```python
## bs객체의 속성들 : 객체변수명.태그명
print(bs.h1)
print("-----------------------")
print(bs.span)
print("-----------------------")
print(bs.div)
print("-----------------------")
print(bs.a)
```

[4] html파일에서 정보 추출 <hr>

```python
## html파일에서 일부분만 선택 str로 저장
data_str = '''<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>BeautifulSoup 활용</title>
</head>'''

## 마크업 STR ==> 객체화
mbs = BeautifulSoup(data_str,'html.parser')
```

```python
## 브라우저 웹 페이지 타이틀 데이터 추출
print("title 태그 :                    ",mbs.title)
print("title 태그에서 속성 추출 :        ",mbs.title.text)
print("title 태그에서 getter메서드로추출 :",mbs.title.get_text())
    
```
