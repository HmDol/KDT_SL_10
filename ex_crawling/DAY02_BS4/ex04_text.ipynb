```python
from urllib.request import urlopen      ## 웹에서 url기반으로 다운로드
from bs4 import BeautifulSoup           ## 웹데이터 객체화 및 정보 추출용 모듈
import pandas as pd

## 데이터 추출
data_path = './fruits-vegetables.html'
data=''
with open(data_path, mode='r', encoding='utf-8') as f :
    data = f.read()

print(type(data) ,data)



```

```python
# bs 객체 생성
mbs = BeautifulSoup(data, 'html.parser')

## - 모든 과일 목록 추출
allF = mbs.find(id = 'fr-list')
print(allF.get_text())

## - 모든 채소 목록 추출
allV= mbs.find(id = 've-list')
print(allV.get_text())
```

```python
## 색상별로 과일 및 채소 목록 추출 : class 속성
fruits = allF.find_all('li')
print(fruits)
color_dict = {}
# for li in fruits :
    # print(li.attrs['class'])
```

```python

df = pd.DataFrame()

# - 모든 과일과 야채를 df에 저장
#     - 이름, 색상, 생산지, 종류 항목 저장
li_all = mbs.find_all('li')
for li in li_all :
    df[li.text] = li.attrs.values()
df = df.T
df = df.reset_index()
df.columns = ['이름','색상','생산지' ]
fruit_list = allF.get_text().split('\n')
df['종류'] = df['이름'].apply(lambda x: '과일' if x in fruit_list else '채소')

display(df)


# - 색상별로 과일 및 야채 목록 추출
# grouped = df.groupby('색상')
# for s,y  in grouped.groups :
#     print(x)
# display(grouped['이름'])

# - 생산지별로 과일 및 야채 목록 추출
```
