### [ 웹 스크래핑 - 정보추출 ]
- 수집된 웹 데이터에서 필요한 정보 추출하는 작업 : 스크래핑
- 스크래핑을 편리하게 진행하기 위한 패키지 : BeautifulSoup4
- binary/str 타입의 웹 데이터 ==> 객체화 (DOM)해서 프로그램 쉽게 사용할 수 있도록 함

[1] 모듈 로딩 <hr>

```python
from urllib.request import urlopen      ## 웹에서 url기반으로 다운로드
from bs4 import BeautifulSoup           ## 웹데이터 객체화 및 정보 추출용 모듈
```

[2] markup 문자열에서 정보 추출<hr>
- 태그사이의 텍스트 읽기
    * 태그 객체명.text : <시작태그> 텍스트만 존재 </끝태그>
    * 태그 객체명.string : <시작태그> 텍스트만 존재 <다른태그> ... </다른태그></끝태그>

- 태그 속성 읽기
    * 태그객체명['속성명]
    * 태그객체명.get('속성명')

```python
## html 파일에서 일부분 선택 str로 저장
html_str = '''
<h1>hello</h1>
<h2>hello<b>world</b></h2>
<a href="http://www.naver.com"> 네이버</a>
<div style="border: 2px solid black; background:#2df941;"> GOOD</div>
'''    
```

```python
## 마크업 str = 객체화
mbs = BeautifulSoup(html_str, 'html.parser')
```

```python
## [1] 태그 전체 가져오기 : bs객체.태그명
print('h1  태그 전체 : ', mbs.h1)
print(' a  태그 전체 : ', mbs.a)
print('div 태그 전체 : ', mbs.div)
```

```python
## [2] 시작 태그 ~ 끝 태그 사이 텍스트 가져오기
print('h1 텍스트 : ', mbs.h1.string)    ## 'hello'  (텍스트만 있어서 ok)
print('h2 텍스트 : ', mbs.h2.string)    ## None     (안에 다른 태그 있어서)
print("-"*30)
## 일반적인 방법 : tag객체명.text 또는 tag객체명.get_text()
print('h2 텍스트 : ', mbs.h2.text, "d\n",mbs.h2.get_text())
print(' a 텍스트 : ', mbs.a.text, mbs.a.get_text())
```

```python
## [3] 태그 속성 가져오기 : tag변수명['속성명'] / tag변수명.get_속성명()
## 

print(' a태그 href 속성값 : ', mbs.a['href'])
print(' a태그 href 속성값 : ', mbs.a.get('href'))

print(' div태그 style 속성값 : ', mbs.div.get('style'))

print(' a태그 class 속성값 : ', mbs.a['class'])
```
