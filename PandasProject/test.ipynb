[1] 컬럼 추출 및 파일 경량화

```python

import pandas as pd

# 원본 파일 경로
file_path = "../Data/한국교통안전공단_중형 전기승용차 센서데이터_20230807.csv"

# 필요한 컬럼 리스트
cols = [
    '속도',                                 ## 차량 주행 속도                  
    '배터리_관리_시스템_고전압팩_전압',         ## HV팩 전류(A)
    '배터리_관리_시스템_고전압팩_전류',         ## HV팩 전류(V)
    '교류_외부_온도',                        ## 외기 온도센서, HVAC제어 및 열관리의 기준
    '교류_컴프레서_압력',                     ## 냉매 회로의 압력(주로 고압측 추정)
    '교류_고전압_히터',                       ## PTC등 전기 히터 출력(가열소자)
    '교류_덕트_온도_드라이버',                 ## 운전석으로 나오는 공조(에어덕트) 공기의 온도
    '클러스터_주행거리'                       ## 클러스터_주행거리
]
df = pd.read_csv(file_path, usecols=cols, encoding='CP949')

# 결측치나 0값 제거
df = df.dropna(subset=['속도', '배터리_관리_시스템_고전압팩_전압', '배터리_관리_시스템_고전압팩_전류'])

# 새로운 경량 파일로 저장
df.to_csv("EV_sensor_core.csv", index=False, encoding='utf-8')

print("경량 데이터 저장 완료: EV_sensor_core.csv")
print("남은 컬럼 수:", len(df.columns))
print("컬럼 목록:", df.columns.tolist())

df.head()
```
