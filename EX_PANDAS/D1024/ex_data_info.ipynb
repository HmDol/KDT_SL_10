[4] 기본 정보 확인 후 컬럼별 자료형 정리 <hr>
- DF.astype('타입명') / SR.astype('타입명') : 원하는 크기 설정 (예 : 'int8', 'float16')
- DF.astype(타입명) : 해당 타입에서 가장 큰 크기 타입 설정 (예 : int64, float64) 

```python
## -------------------------------------------------------
## 형변환 컬럼 정리
## - horsepower : object --> int / '?' 처리
## - cylinders  : int64 -> category
## - origin     : int64 -> category
## -------------------------------------------------------

## 1) horsepower 컬럼
## - '?' 개수 파악
display(mpgDF.horsepower.value_counts()['?']) # value_counts : 시리즈반환! 

## - ? 데이터 존재하는 행 삭제
ret = mpgDF.horsepower == '?'
del_index = mpgDF[ret].index
mpgDF2 = mpgDF.drop(index=del_index) ## 원본 유지
print(mpgDF2.shape)

## object ==> int16
mpgDF2.horsepower = mpgDF2.horsepower.astype('int16') ## inplace없음. 반드시 저장 필요

## - 행 삭제 후 인덱스 정리 필요
mpgDF2.reset_index(drop=True, inplace=True)

## 정보 확인
mpgDF2.info()
```

```python
## 2) cylinders  : int64 -> category
##    origin     : int64 -> category
mpgDF2[['cylinders','origin']] = mpgDF2[['cylinders','origin']].astype('category')

## 확인
mpgDF2.info()
```
