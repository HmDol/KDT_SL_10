### 【 데이터 그룹화 연산 】


[1] 모듈 로딩 및 데이터 준비<hr>


```python
## 모듈 로딩
import pandas as pd
```

```python
## 데이터 준비
DATA_FILE = '../Data/titanic.csv'

## CSV => DataFrame 로딩
titanicDF = pd.read_csv(DATA_FILE, usecols=[0,2,3,6,8])
titanicDF = pd.read_csv(DATA_FILE, usecols=['survived',	'sex',	'age',	'fare',	'class'])
```

```python
titanicDF
```

[2] 데이터 기본 정보 확인 : info(), head(), tail(), describe() <hr>

```python
## 전체 데이터 요약 정보
titanicDF.info()

## 실제 데이터 확인
display(titanicDF.head(2), titanicDF.tail(2))

## 모든 컬럼별 통계값 확인
display(titanicDF.describe(include ='all'))
```

```python
## 기본 정보 확인 후 파악된 내용
## - age 컬럼 결측치가 많음
## - survived, sex, class 컬럼이 가진 데이터의 의미 => 범주형 category
## - 가로/행 의미 즉, 1명 생존자 정보이므로 중복 허용 여부 => 5개 컬럼으로는 unique x 중복 허용 가능
```

[3] 그룹화 진행 <hr>

```python
##[3-1] class 컬럼으로 그룹화
classDGB = titanicDF.groupby(['class'], as_index=False)

## 그룹화 후 속성들
print(f'그룹 수 : {classDGB.ngroups}')
for k in classDGB.groups :
    print(k)
    print(classDGB.get_group(k).head())

print(f'그룹별 데이터 행수 \n{classDGB.size()}')
```

```python
## [3-2] 그룹별 집계
display(classDGB.mean(numeric_only=True))

display(classDGB.sum(numeric_only=True))

display(classDGB.std(numeric_only=True))

display(classDGB[['survived','age','fare']].agg(['sum', 'mean']))
```

```python
display( classDGB['fare'].sum(numeric_only=True))
display( classDGB['fare'].transform('sum'))

## 그룹별 연산 결과를 동일 DF/Series로 반환
titanicDF['total'] = classDGB['fare'].transform('sum')
titanicDF.head()
```
