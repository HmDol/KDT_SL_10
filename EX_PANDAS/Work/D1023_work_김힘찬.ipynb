```python
## 문제 1. 아래 데이터를 저장 후 조건을 만족하는 코드를 작성하세요.

import pandas as pd

data = [
    ["2차전지(생산)", "SK이노베이션", 10.19, 1.29],
    ["해운", "팬오션", 21.23, 0.95],
    ["시스템반도체", "티엘아이", 35.97, 1.12],
    ["해운", "HMM", 21.52, 3.20],
    ["시스템반도체", "아이에이", 37.32, 3.55],
    ["2차전지(생산)", "LG화학", 83.06, 3.75] 
] 
cols = ['테마', '종목명', 'PER', "PBR"]

df = pd.DataFrame(data, columns=cols)
# display(df)

## (1-1) 종목명, PER, PBR 컬럼만 선택 후 DataFrame으로 저장하는 코드 작성하세요.
df1 = df[['종목명', 'PER', 'PBR']]
print(df1)

## (1-2) PER이 20미만, 50이상인 데이터만 저장하는 코드 작성하세요.
ret = (df['PER'] < 20) | (df['PER'] >= 50)
df[ret]

## (1-3) 종목명을 행인덱스롤 설정하는 코드 작성하세요. 
df = df.set_index('종목명')

# (1-4) PBR 컬럼의 데이터를 기준으로 내림차순 정렬해 주세요. 
df.sort_values(by='PBR', ascending=False, inplace=True)

# (1-5) PER,, PBR, PER 순서로 오름차순 정렬해 주세요.
df.sort_values(by=['PER', 'PBR'], ascending=[True, True], inplace=True)


display(df)
```

```python
## 문제2. 아래 데이터를 저장 후 조건을 만족하는 코드를 작성하세요. 
data = [
 ["037730", "3R", 1510, 7.36],
 ["036360", "3SOFT", 1790, 1.65],
 ["005760", "ACTS", 1185, 1.28], 
] 
columns = ["종목코드", "종목명", "현재가", "등락률"] 

df = pd.DataFrame(data,columns=columns)
display(df)

## (2-1) "종목코드"를 행 인덱스로 설정하세요. 
df = df.set_index('종목코드')
print(df)

## (2-2) 아래 결과가 출력되도록 코드 작성하세요. 
df1 = df.iloc[0:2, 0:2]
print(df1)

## (2-3) 현재가가 1,500원 이상이면서 1,700원 미만인 값만 출력하세요.
ret = (df["현재가"] >= 1500) & (df["현재가"] < 1700)
print(df[ret])

## (2-4) 목표가는 현재가에서 10% 상승한 가격으로 목표가 컬럼을 추가하세요. 
df["목표가"] = df["현재가"] * 1.1
display(df)

## (2-5) LG 데이터 추가
new_row = pd.DataFrame(
    [["LG전자", 60000, 3.84, 60000 * 1.1]],
    columns=["종목명", "현재가", "등락률", "목표가"],
    index=["066570"]
)
df = pd.concat([df, new_row])
display(df)
```

```python
# 문제 3. 아래 데이터를 저장 후 조건에 맞도록 코드를 작성하세요. 
import numpy as np 
data = {
    "카카오": [np.nan, 221, 35],
    "NAVER": [52, 47, 3]
    } 
index = [2019, 2020, 2021] 
df = pd.DataFrame(data, index=index )

# print(df)

##(3-1) NAVER 컬럼 삭제해 주세요. 
df = df.drop(columns="NAVER")
# display(df)

## (3-2) "삼성전자": [17, 21, 13] 데이터를 추가해 주세요. 
df["삼성전자"] = [17, 21, 13]
# print(df)

## (3-3) 컬럼명을 기준으로 내림차순 정렬해 주세요.
df = df.sort_index(axis=1, ascending=False)
# print(df)

## (3-4) 아래 형태가 되도록 코드 작성 하세요 
#       2019 2020 2021 
# 카카오 NaN 221.0 35.0 
# NAVER 52.0 47.0 3.0
df = df.T
display(df)
```

```python
# 문제 4. 아래 데이터를 저장 후 조건을 만족하는 코드를 작성하세요. - 데이터 파일 : 'vgsales.csv'  
FILE_NAME = '../Data/vgsales.csv'
df = pd.read_csv(FILE_NAME)
# display(df)
# (4-1) 'Rank' 컬럼을 인덱스로 설정하면서 DataFrame으로 변환하세요.
df = df.set_index('Rank')
# display(df)
# (4-2) 'Global_Sales' 컬럼을 기준으로 내림차순 정렬하세요.
df = df.sort_values(by='Global_Sales', ascending=False)
# display(df)

# (4-3) 'Platform'과 'Global_Sales'를 기준으로 정렬하세요.
df = df.sort_values(by=['Platform', 'Global_Sales'], ascending=[True, False])
# display(df)

# (4-4) 'Year' 컬럼의 타입을 int로 변경하세요.
# 'Platform'은 오름차순으로 정렬하고 'Global_Sales'는 내림차순으로 정렬 
# # 결측값(NaN) 제거 또는 처리 후 변환해야 오류 안 남
df = df.dropna(subset=['Year'])
df['Year'] = df['Year'].astype(int)
# print(df['Year'].dtypes)  # int로 변환 확인

# (4-5) 게임 이름(Name)에 'Super Mario'가 포함된 게임들만 선택하세요. 
mario_df = df[df['Name'].str.contains('Super Mario', case=False, na=False)]
display(mario_df)
```
