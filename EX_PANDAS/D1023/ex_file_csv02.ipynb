### [ 파일 입력과 출력 ]<hr>
- pandas.read_xxx() : 다양한 포맷의 파일 ==> dataFrame으로 변환
    * 파일 종류에 따라서 변환 시 설정 필요함

- pandas.to_xxx() : DataFrame ===> 다양한 포맷 str 또는 file로 변환

```python
## [1] 모듈 로딩
import pandas as pd         # 데이터분석 모듈
import os                   # 파이썬 내장 모듈, 경로, 폴더, 시스템 관련 처리 모듈
import glob                 # 유닉스 운영체제 스타일의 파일, 폴더 경로 처리 모듈
```

```python
## [2] 데이터 준비 - 일반 csv 파일
DATA_DIR = '../Data'

## 폴더내 파일명을 리스트에 담아서 반환
FILE_LIST = os.listdir(DATA_DIR)
print(FILE_LIST)

## 파일경로 + 파일명
FILE_NAME = f'{DATA_DIR}/{FILE_LIST[0]}'

print(FILE_NAME)
```

```python
## [2] 데이터 준비 - 일반 csv 파일

## 폴더내 파일명을 리스트에 담아서 반환
## *.* 의미 : 모든 파일명.모든확장자
FILE_LIST = glob.glob(f'{DATA_DIR}/*.*')
print(FILE_LIST)

FILE_LIST = glob.glob(f'{DATA_DIR}/*.csv')
print(FILE_LIST)

FILE_LIST = glob.glob(f'{DATA_DIR}/*.xml')
print(FILE_LIST)

FILE_LIST = glob.glob(f'{DATA_DIR}/*.csv')
print(FILE_LIST)
```

```python
## [3] csv --> df으로 변환
FILE_NAME = f'{DATA_DIR}/{FILE_LIST[0]}'
## 앞부분 일부 데이터 제거하고 가져오기 : skiprows = 51
csvDF = pd.read_csv(FILE_NAME, skiprows=51, header=None)

## 출력
csvDF
```

```python
## [3] csv --> df으로 변환

## 끝부분 일부 데이터 제거하고 가져오기 : skipfooter = 50
##                   추가로 엔진설정 : engine='python'
csvDF = pd.read_csv(FILE_NAME, skipfooter=50, engine='python')

## 출력
csvDF
```

```python
## [5] 데이터 준비 - 특정 컬럼을 행인덱스 설정
FILE_NAME = '../Data/test2.csv'

## 날짜/시간 str로 object 타입 ==> dataetime64[ns] 타입 변환
csvDF = pd.read_csv(FILE_NAME, sep=" ", parse_dates=['DATE'])

print(csvDF, csvDF.dtypes, sep="\n\n")
```

```python
## [6] 데이터 준비 - 일부 컬럼만 DataFrame으로 저장
FILE_NAME = '../Data/test.csv'

## 원하는 컬럼만 가져오기
## usescols 매개변수 -> usescols = ['컬럼명1', '컬럼명2']
##                  -> usescols = [정수, 정수]

csvDF = pd.read_csv(FILE_NAME, usecols=['c1', 'c2'])
csvDF2 = pd.read_csv(FILE_NAME, usecols=[1,2])


## 출력
display(csvDF, csvDF2, sep="\n\n")
```
