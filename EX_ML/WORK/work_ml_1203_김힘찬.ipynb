##### 【 ML - 꽃 품종 식별 모델 구현 】

- 주____제 : 3개 꽃 품종 식별 모델 구현으로 꽃가게 웹 서비스 
- 데이터셋
    * 파일명: Data/iris.csv
    * 피쳐들: **꽃받침 길이, 꼭받침 너비, 꽃잎 길이, 꽃잎 너비**
    * 타__겟: **분꽃 종류**
- 학습_종류: **지도학습 => 분류**
- 학습_방법: **KNN 최근접 이웃 알고리즘**

- 조____건  
    (1) 4개의 피쳐를 모두 사용한 모델  
    (2) 4개중 중요 피쳐만 사용한 모델

[1] 모듈 로딩 <hr>

```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
```

```python
## 데이터 로딩
data_file = '../Data/iris.csv'

df = pd.read_csv(data_file)
display(df.head())

## 결측치 / 이상치 확인
df.info()

for x in df.columns :
    print(f'[{x}] -------------- {df[x].nunique()}개')
    print(f'{df[x].unique()}')
    
```

```python
## 상관계수 확인
df.corr(numeric_only=True)
```

```python
## 모델1 : 4개의 피쳐 모두 사용
## 피쳐 데이터
featureDF = df[df.columns[:-1]]

## 타겟 데이터 
targetSR = df[df.columns[-1]]

print(f'featureDF : {featureDF.ndim}D, targetSR : {targetSR.ndim}D, ')

## 모델 인스턴스 생성
m1 = KNeighborsClassifier()
m1

## 학습 진행
m1.fit(featureDF, targetSR)

## 평가 진행
m1.score(featureDF, targetSR)
```

```python
## 모델2 : 4개중 중요 피쳐만 사용
## 피쳐 데이터
featureDF = df[df.columns[2:-1]]

## 타겟 데이터 
targetSR = df[df.columns[-1]]

print(f'featureDF : {featureDF.ndim}D, targetSR : {targetSR.ndim}D, ')

## 모델 인스턴스 생성
m2 = KNeighborsClassifier()
m2

## 학습 진행
m2.fit(featureDF, targetSR)

## 평가 진행
m2.score(featureDF, targetSR)
```
