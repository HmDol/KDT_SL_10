##### [ 이미지 변형 - & ROI ]
- 픽셀 연산 수행 이미지 변화
    * 이미지 데이터 => 행렬, numpy의 ndarray
    * 덧__셈 연산 => 각각의 픽셀 단위 덧셈
    * 뺄__셈 연산 => 각각의 픽셀 단위 뺄셈
    * 곱__셋 연산 => 각각의 픽셀 단위 곱셋
    * 나눗셈 연산 => 각각의 픽셀 단위 나눗셈
- 색상값 : 0 검정색 black - 1 또는 255 흰색 white
    * 0 >>> 1 또는 255 : 밝아짐
    * 0 <<< -1 또는 255 : 어두어짐
- 이미지 밝기 조절
- 이미지 자연스로운 합성

[1] 이미지 픽셀 기본 연산 <HR>

```python
## 모듈 로딩
import numpy as np
import cv2
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
## [1-2] 데이터 준비
img_file = '../Data/Image/ddd.png'

oriNP = cv2.imread(img_file)
oriNP = cv2.cvtColor(oriNP, cv2.COLOR_BGR2RGB)
grayNP = cv2.imread(img_file, cv2.IMREAD_GRAYSCALE)
grayNP = cv2.cvtColor(grayNP, cv2.COLOR_BGR2RGB)

## 이미지 기본정보 확인
print(f'oriNP 정보 : {oriNP.shape}') if oriNP is not None else print("이미지 없음")
print(f'grayNP 정보 : {grayNP.shape}') if grayNP is not None else print("이미지 없음")
```

[3] 이미지 밝기 조절 <hr>

```python
##[3-1] 컬러 이미지 밝기 조절
### 0 >>>> 255 : 흰색, white 밝아짐
colorNP30 = cv2.add(oriNP, 30)
colorNP50 = cv2.add(oriNP, 50)
colorNP100 = cv2.add(oriNP, 100)
```

```python
## 시각화
fig, axes = plt.subplots(1,4, figsize = (20,5))
numobj = [oriNP, colorNP30, colorNP50, colorNP100]
titles = ['ori', '30', '50', '100']

for ax, img, title in zip(axes, numobj, titles) :
    ax.imshow(img)
    ax.set_title(title)
    ax.set_axis_off()
plt.show()
```

```python
### 0 <<<< 255 : 흰색, white 밝아짐
colorNP30 = cv2.subtract(oriNP, 30)
colorNP50 = cv2.subtract(oriNP, 50)
colorNP100 = cv2.subtract(oriNP, 100)
```

```python
## 시각화
fig, axes = plt.subplots(1,4, figsize = (20,5))
numobj = [oriNP, colorNP30, colorNP50, colorNP100]
titles = ['ori', '30', '50', '100']

for ax, img, title in zip(axes, numobj, titles) :
    ax.imshow(img)
    ax.set_title(title)
    ax.set_axis_off()
plt.show()
```

```python
##[3-1] 컬러 이미지 밝기 조절
### 0 >>>> 255 : 흰색, white 밝아짐
colorNP30 = cv2.add(grayNP, 30)
colorNP50 = cv2.add(grayNP, 50)
colorNP100 = cv2.add(grayNP, 100)
```
