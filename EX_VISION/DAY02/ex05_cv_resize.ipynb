#### [openCV - 이미지 크기 변경]

- 이미지의 확대/축소 기능
    - 함수 : cv2.resize(로우이미지 데어터, dsSize=(w,h), fx=상대크기, fy=상대크기, interporation = 보간법) 
    - 결과 : 변경된 크기의 이미지 로우 데이터
    
- 주의 : opencv 한글 미지원

```python
## 이미지 처리 모듈
import cv2

##시각화 모듈
import matplotlib.pyplot as plt
# %pip install koreanize_matplotlib
import koreanize_matplotlib
```

```python
img_file = '../Data/Image/ddd.png'
# img_file = '../Data/Image/flower_rainbow.jpg'
```

[2] 이미지 로딩 및 정보 확인 <hr>

```python
## 원본 & 회색 로딩
org_img = cv2.imread(img_file)
gray_img = cv2.imread(img_file, cv2.IMREAD_GRAYSCALE)
```

```python
## 이미지 데이터 정보 확인
print(f'[orinal image]')
print(f'형태 : {org_img.shape} / 차원 : {org_img.ndim}D / 사이즈 : {org_img.size}')

print(f'[gray image]')
print(f'형태 : {gray_img.shape} / 차원 : {gray_img.ndim}D / 사이즈 : {gray_img.size}')
```

```python
## 이미지 시각화
fig, axes = plt.subplots(1,2, sharey=True)
print(axes)

axes[0].imshow(org_img)
axes[0].set_title('컬러 이미지')
axes[0].set_xticklabels([])
axes[0].set_yticklabels([])


axes[1].imshow(gray_img, cmap='grey')
axes[1].set_title('흑백 이미지')
axes[1].set_xticklabels([])
axes[1].set_yticklabels([])
# plt.imshow(org_img)
plt.suptitle("이미지 시각화")
plt.show()
```

[3] 이미지 확대/수축 <hr>

```python
h, w, _ = org_img.shape

org_rgb = cv2.cvtColor(org_img, cv2.COLOR_BGR2RGB)

down = cv2.resize(org_rgb, (int(h/3), int(h/3)), interpolation=cv2.INTER_AREA)
# down_rgb = cv2.cvtColor(down, cv2.COLOR_BGR2RGB)

up = cv2.resize(org_rgb, (int(h*3), int(h*3)), interpolation=cv2.INTER_CUBIC)
# up_rgb = cv2.cvtColor(up, cv2.COLOR_BGR2RGB)

imgs = [org_img[:,:,::-1], down, up]
titles = [f"orig {org_rgb.shape}", f"down {down.shape}", f"up {up.shape}"]

fig, axes = plt.subplots(1,3, sharey=True)
for idx, ax in enumerate(axes):
    ax.imshow(imgs[idx])
    ax.set_title(titles[idx])
    ax.set_axis_off()
plt.tight_layout()
plt.show()
```
