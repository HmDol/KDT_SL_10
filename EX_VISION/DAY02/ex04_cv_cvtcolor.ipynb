#### [openCV - 이미지 색공간(color space) 변경]

- 목적에 따른 색상 표현해 주는 색공간 변경을 위한 것
    - 함수 : cv2.cvtColor(로우이미지데이터, code)
        * code 매개변수 : cv2.COLOR_BGR2RGB, cv2.COLOR_BGR2HSV, ...
    - 결과 : 튜플 타입에 분리된 채널 데이터 반환

- 채널 병합
    - 함수 : cv2.merge(채널데이터, 채널 데이터, 채널 데이터)
    - 결과 : 합쳐진 이미지 로우 데이터

- 주의 : opencv 한글 미지원

```python
## 이미지 처리 모듈
import cv2

##시각화 모듈
import matplotlib.pyplot as plt
# %pip install koreanize_matplotlib
import koreanize_matplotlib
```

```python
img_file = '../Data/Image/ddd.png'
# img_file = '../Data/Image/flower_rainbow.jpg'
```

[2] 이미지 로딩 및 정보 확인 <hr>

```python
## 원본 & 회색 로딩
org_img = cv2.imread(img_file)
gray_img = cv2.imread(img_file, cv2.IMREAD_GRAYSCALE)
```

```python
## 이미지 데이터 정보 확인
print(f'[orinal image]')
print(f'형태 : {org_img.shape} / 차원 : {org_img.ndim}D / 사이즈 : {org_img.size}')

print(f'[gray image]')
print(f'형태 : {gray_img.shape} / 차원 : {gray_img.ndim}D / 사이즈 : {gray_img.size}')
```

```python
## 이미지 시각화
fig, axes = plt.subplots(1,2, sharey=True)
print(axes)

axes[0].imshow(org_img)
axes[0].set_title('컬러 이미지')
axes[0].set_xticklabels([])
axes[0].set_yticklabels([])


axes[1].imshow(gray_img, cmap='grey')
axes[1].set_title('흑백 이미지')
axes[1].set_xticklabels([])
axes[1].set_yticklabels([])
# plt.imshow(org_img)
plt.suptitle("이미지 시각화")
plt.show()
```

[3] 이미지 색공간 변경 <hr>

```python
## [3-1] BGR -> RGB
img_rgb = cv2.cvtColor(org_img, cv2.COLOR_BGR2RGB)
planes = cv2.split(img_rgb)
plt.imshow( img_rgb)
plt.show()

```

```python
## [3-2] BGR => HSV
img_hsv = cv2.cvtColor(org_img, cv2.COLOR_BGR2HSV)
planes = cv2.split(img_hsv)
plt.imshow(img_hsv)
plt.show()

```

```python
## [3-3] BGR => ?
img_hls = cv2.cvtColor(org_img, cv2.COLOR_BGR2HLS)
planes = cv2.split(img_hls)
plt.imshow(img_hls)
plt.show()
```

```python
## 시각화
titles = ['origin(bgr)', 'rgb', 'hsv', 'hls']
rawimg = [org_img, img_rgb,img_hsv, img_hls]

fig, axes = plt.subplots(1,4, sharey =True)
for idx ,ax in enumerate(axes) :
    ax.imshow(rawimg[idx], cmap='grey')
    ax.set_title(titles[idx])
    ax.axis('off')
plt.tight_layout()
plt.show()
```
