#### python & db 연결

```python
import pymysql

print(f'pymysql v.{pymysql.__version__}')
```

```python
## RDBMS 정보
server_ip = 'localhost'
user_id = 'root'
user_pw = 'mysql'
db_name = 'mydb'
charset = 'utf8mb4'
```

```python
## ========================================================
## db 연결
## ========================================================
conn = pymysql.connect(host = server_ip,
                user = user_id,
                password = user_pw,
                database= db_name,
                cursorclass = pymysql.cursors.DictCursor)
```

```python
## ========================================================
## db - select 처리
## ========================================================
all_data_sql = 'select * from user;'

with conn.cursor() as cursor :
    ## 모든 데이터 다 가져오기
    # sql = 'select id, name from user where name = %s'
    cursor.execute(all_data_sql)
    result = cursor.fetchall()
    print(result)
```

```python
## ========================================================
## db - insert 처리
## ========================================================
isnert_data_sql = "INSERT INTO user VALUES (%s,%s,%s,%s,%s,%s);"

with conn.cursor() as cursor :
    cursor.execute(isnert_data_sql, (29,  '베트맨', 23, '부산', '부산시', '여'))
    conn.commit()
```

```python
## ========================================================
## db - delete 처리
## ========================================================
delete_data_sql = "delete from user where id=%s"

with conn.cursor() as cursor :
    cursor.execute(delete_data_sql, (29,))
    conn.commit()
```

```python
## ========================================================
## db 연결 해제
## ========================================================
conn.close()
```
