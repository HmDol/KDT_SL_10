## [Matpotlib의 pie 그래프]
- 전체 대비 항목에 대한 비중/비율 시각하 그래프
- 데이터 제공시 자동 계산 수행


```python
## 모듈 로딩
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
## 데이터 준비
labels = 'Frog', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]

## 시각화 [기본]
plt.pie(sizes, labels=labels)
plt.show()
```

```python
## -------------------------------------------------------
## 조각별 비율 출력 => autopct 매개변수 : 00.00%
## -------------------------------------------------------
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()
```

```python
## -------------------------------------------------------
## 조각별 비율 출력 + 색 설정
## -------------------------------------------------------
plt.pie(sizes, labels=labels, autopct='%1.1f%%',
        colors=['orange', 'lightblue', 'lightgray', 'yellow'])
plt.title('[항목별 비율]')
plt.show()
```

```python
## -------------------------------------------------------
## 라벨/퍼센트값 위치 조절 => pctdistance / labeldistance 매개변수
##              [기본값] :  [0.6]         [1.1]
## -------------------------------------------------------
plt.pie(sizes, labels=labels, autopct='%1.1f%%',
        colors=['orange', 'lightblue', 'lightgray', 'yellow'],
        pctdistance=1.2, labeldistance=0.6)
plt.title('[항목별 비율]')
plt.show()
```

```python
## -------------------------------------------------------
## 조각 간격 주기 => explode 매개변수
## 그림자 효과 설정 => shadow 매개변수
## -------------------------------------------------------
plt.pie(sizes, labels=labels, autopct='%1.1f%%',
        colors=['orange', 'lightblue', 'lightgray', 'yellow'],
        pctdistance=1.2, labeldistance=0.6,
        explode = (0.1,0,0,0.4), shadow=True)
plt.title('[항목별 비율]')
plt.show()


```

```python
## -------------------------------------------------------
## 시작 각도 설정   => startangle  = 0 [기본]
## 방향            => countclock
## 그림자 효과 설정 => shadow 매개변수
## -------------------------------------------------------
plt.pie(sizes, labels=labels, autopct='%1.1f%%',
        colors=['orange', 'lightblue', 'lightgray', 'yellow'],
        pctdistance=1.2, labeldistance=0.6,
        explode = (0.1,0,0,0.4), shadow=True,
        startangle = 180, counterclock=False)
plt.title('[항목별 비율]')
plt.show()
```

```python
## -------------------------------------------------------
## 도넛 그래프 : pie그래프 변형
## - 쐐기 wedge 매개변수 설정 dict 타입
## ------------------------------------------------------w
wedgedict={'width': 0.6, 'linewidth':0.5, 'edgecolor':'white'}
plt.pie(sizes, labels=labels, autopct='%1.1f%%',
        colors=['orange', 'lightblue', 'lightgray', 'yellow'],
        pctdistance=1.2, labeldistance=0.6,
        startangle = 180, counterclock=False,
        wedgeprops =wedgedict,
        explode=[0.1,0.01,0.01,0.1])

plt.title('[항목별 비율]')
plt.show()
```

```python
## -------------------------------------------------------
## 범례 : pie(labels) 값을 항목으로 표시
## -> loc='best' : upper left, upper left, lower left ......
## ------------------------------------------------------w
wedgedict={'width': 0.6, 'linewidth':0.5, 'edgecolor':'white'}
plt.pie(sizes, labels=labels, autopct='%1.1f%%',
        colors=['orange', 'lightblue', 'lightgray', 'yellow'],
        pctdistance=1.2, labeldistance=0.6,
        startangle = 180, counterclock=False,
        wedgeprops =wedgedict,
        explode=[0.1,0.01,0.01,0.1])

## 현재 그래프 figure, axes
fig = plt.gcf()
fig.set_facecolor('lightyellow')

plt.legend(bbox_to_anchor=(1,1))
plt.title('[항목별 비율]')
plt.show()
```
