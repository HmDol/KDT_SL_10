```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
## 데이터 준비
FILE_PAHT = '../Data/남북한발전전력량.xlsx'

dataDF = pd.read_excel(FILE_PAHT)
display(dataDF)
dataDF.info()

```

```python
## 컬럼명 변환
dataDF = dataDF.rename(columns={'전력량 (억㎾h)' : '남/북'})

## 남/북 컬럼 결측치 채우기
dataDF['남/북'] = dataDF['남/북'].ffill()


## '-' --> 0으로 판단
dataDF = dataDF.replace('-', 0)


## 북한 신재생 행 추가하기
new_row = {'남/북' : '북한', '발전 전력별' : '신재생'}
# 0으로 채우기
for col in dataDF.columns :
    if col not in new_row :
        new_row[col] = 0
print(new_row)
dataDF.loc[len(dataDF)] = new_row
display(dataDF.head() ,dataDF.tail())
```

```python
## 멀티 인덱스 
dataDF = dataDF.set_index(['남/북', '발전 전력별'])
dataDF.index
```

```python
dataDF['항목별총합'] = dataDF.sum(axis=1)
dataDF.loc['납북한합계'] = dataDF.sum(axis=0)

dataDF.tail()
```

```python
for idx1, idx2 in dataDF.index[1:5] :
    plt.plot(dataDF.loc[idx1, idx2], label=idx2)
    print(idx2)
       
```
