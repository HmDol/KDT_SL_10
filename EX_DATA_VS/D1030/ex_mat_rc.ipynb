```python
## 모듈 로딩
from matplotlib import rcParams
import pandas as pd
```

[2] 마이너스 기호 설정<hr>
- 기본값 : 유니코드 '-' (U+2212)
- 문제점 : 한글폰트에 해당 코드가 없음 => □(깨짐)으로 출력


```python
## [설정방법 1]
from matplotlib import rcParams

## window 유니코드 '-' 대신 ==> ascii '-' 사용설정
rcParams['axes.unicode_minus'] = False
```

```python
## [설정방법 2]
import matplotlib.pyplot as plt

plt.rcParams['axes.unicode_minus'] = False
```

[3] Font 설정<hr>

```python
## 폰트 관련 모듈
from matplotlib import font_manager as fm
import matplotlib

## - 시스템 폰트 정보
fm.findSystemFonts()
```

```python
## 폰트 파일경로 + 파일명
WINDOW_SYS_FONT=r'C:\\Windows\\Fonts\\HTOWERT.TTF'
FONT_PATH = './MALGUN.TTF'

## 폰트 인스턴스 생성 후 이름 추출
font_name = fm.FontProperties(fname=WINDOW_SYS_FONT).get_name()

## 폰트 이름 등록 및 설정
matplotlib.rc('font', family=font_name)
```
