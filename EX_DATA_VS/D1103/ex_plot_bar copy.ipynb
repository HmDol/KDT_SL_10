### 【 데이터 시각화 - Area Graph 】
- 선 그래프에서 추가된 기능
- 특정 영역을 가종/집중해서 표현하고 싶은 경우 사용
- 너무 많은 조각은 오히려 가독성 및 집중도 하락
- 함수 : matplotlib.pyplot.fill_between()/fill_betweenx()

```python
### [1] 모듈 로딩 
import matplotlib.pyplot as plt
import koreanize_matplotlib 
import pandas as pd
```

```python
## ===========================================================
## [2] 선 그래프
## ===========================================================

## - 데이터
xData = [1,2,3,4]
yData = [2,3,5,10]

line = plt.plot(xData, yData)

## 그래프 기본요소
plt.title('기본 선 그래프')
plt.xlabel('x 데이터')
plt.ylabel('y 데이터')
plt.grid()              ##눈금선

## 화면에 그래프 출력
plt.show()
```

```python
## ===========================================================
## [2] 선 그래프 - 영역 설정
## ===========================================================

## - 데이터
xData = [1,2,3,4]
yData = [2,3,5,10]

line = plt.plot(xData, yData)

## 특정 영역 선 아래 채우기
## - fill_between(x, y) : x축 기준
plt.fill_between(xData[1:3],yData[1:3], alpha=0.2)

## 그래프 기본요소
plt.title('기본 선 그래프')
plt.xlabel('x 데이터')
plt.ylabel('y 데이터')
plt.grid()              ##눈금선

## 화면에 그래프 출력
plt.show()
```

```python
## ===========================================================
## [2] 선 그래프 - 영역 설정
## ===========================================================

## - 데이터
xData = [1,2,3,4]
yData = [2,3,5,10]

line = plt.plot(xData, yData)

## 특정 영역 선 아래 채우기
## - fill_betweenx(y, x) : y축 기준
plt.fill_betweenx(yData[2:4],xData[2:4], alpha=0.2)

## 그래프 기본요소
plt.title('기본 선 그래프')
plt.xlabel('x 데이터')
plt.ylabel('y 데이터')
plt.grid()              ##눈금선

## 화면에 그래프 출력
plt.show()
```

```python
## ===========================================================
## [2] 선 그래프 + 영역설정 : 그래프에서 집중영역 설정
## ===========================================================

## - 데이터
xData = [1,2,3,4]
yData = [2,3,5,10]
yData2 = [1,2,4,8]

line = plt.plot(xData, yData)
line2 = plt.plot(xData, yData2)
## 특정 임의 영역 채우기
plt.fill([1.9, 1.9, 3.1, 3.1], [1.0, 4.0, 6.0, 3.0], color='lightgray', alpha=0.5)

## 그래프 기본요소
plt.title('기본 선 그래프')
plt.xlabel('x 데이터')
plt.ylabel('y 데이터')
plt.grid()              ##눈금선

## 화면에 그래프 출력
plt.show()
```
