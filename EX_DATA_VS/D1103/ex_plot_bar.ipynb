### 【 데이터 시각화 - 공통사항 】
- 데이터 비교 시 많이 사용되는 그래프
- 정렬 시켜서 시각화시 빠르게 인식 가능
- 데이터따라서 수평/수직으로 설정해서 시각하ㅗ
- 함수 : matplotlib.pyplot.bar()/barh()

```python
### [1] 모듈 로딩 
import matplotlib.pyplot as plt
import koreanize_matplotlib 
import pandas as pd
```

```python
## [2] 기본 막대 그래프

## - 데이터
xData = [10,20,30,40,50]
yData = [10,20,30,40,50]

bar = plt.bar(range(1,len(xData)+1), xData)

## 그래프 기본요소
plt.title('기본 막대 그래프')
plt.xlabel('x 데이터')
plt.ylabel('y 데이터')


## 화면에 그래프 출력
plt.show()
```

```python
## ---------------------------------------------------------------------
## [2] 기본 막대 그래프 - dataFrame
## ---------------------------------------------------------------------

## - 데이터
dataDF = pd.DataFrame({'A' : [10,50,30,40,20],
                        'B' : [15,35,45,25,55]})
display(dataDF)

dataDF = dataDF.sort_values(by=['A'], ascending=False).reset_index(drop=True)
display(dataDF)

## - 막대 그래프 인스턴스 생성 : 기본 값으로 선 그래프 생성
## - align : 막대와 라벨 위치 설정
## - width : 막대 너비 설정 [기] 0.8
## - color : 막대 색상 설정
## 색상코드 값
colors = ["#ff0000", "#0062ff", "#ee00ff", "#00ff3c", "#9000ff", "#e5ff00"]
bar = plt.bar(dataDF.A.index, dataDF.A, 
              align='edge', width=0.5,color=colors)

## 그래프 기본요소
plt.title('기본 막대 그래프')
plt.xlabel('x 데이터')
plt.ylabel('y 데이터')


## 화면에 그래프 출력
plt.show()
```

```python
## ---------------------------------------------------------------------
## [3] 기본 막대 그래프 - dataFrame + 수평
## ---------------------------------------------------------------------

## - 데이터
dataDF = pd.DataFrame({'A' : [10,50,30,40,20],
                        'B' : [15,35,45,25,55]})


## 컬럼별 합계
sumSR = dataDF.sum()


## - 막대 그래프 인스턴스 생성 : 기본 값으로 선 그래프 생성
## - align : 막대와 라벨 위치 설정
## - width : 막대 너비 설정 [기] 0.8
## - color : 막대 색상 설정
## 색상코드 값
colors = [ "#ee00ff", "#00ff3c"]


## width(수직) <-> height(수평)
bar = plt.bar(sumSR.index, sumSR.values, align='center', width=0.3,color=colors)


## 그래프 기본요소
plt.title('컬럼별 합계')
plt.xlabel('컬럼')
plt.ylabel('합계')
3
## 화면에 그래프 출력
plt.show()
```

```python
## ---------------------------------------------------------------------
## [3] 응용 막대 그래프
## ---------------------------------------------------------------------

## - 데이터
dataDF = pd.DataFrame({'A' : [10,50,30,40,20],
                        'B' : [15,35,45,25,55]})

## 컬럼별 합계
sumSR = dataDF.sum()


## - 막대 그래프 인스턴스 생성 : 기본 값으로 선 그래프 생성
## - align : 막대와 라벨 위치 설정
## - width : 막대 너비 설정 [기] 0.8
## - color : 막대 색상 설정
## 색상코드 값
colors = [ "#ee00ff", "#00ff3c"]


## width(수직) <-> height(수평)
bar = plt.bar(sumSR.index, sumSR.values, align='edge', width=0.3,color=colors)

print(bar.patches)
for rect in bar : 
    # print(rect.get_height(), rect.get_width(), rect.get_x())
    h,w,x = rect.get_height(), rect.get_width(), rect.get_x()
    plt.text(x+(w/2),h/2,f'합계:{h}',fontdict={'fontsize' : 10, 'color' : 'blue'},
             ha='center', va='center')

## 그래프 기본요소
plt.title('컬럼별 합계')
plt.xlabel('컬럼')
plt.ylabel('합계')


## 화면에 그래프 출력
plt.show()
```

```python
## ---------------------------------------------------------------------
## [3] 응용 막대 그래프
## ---------------------------------------------------------------------

## - 데이터
dataDF = pd.DataFrame({'A' : [10,50,30,40,20],
                        'B' : [15,35,45,25,55]})

## 컬럼별 합계
sumSR = dataDF.sum()


## - 막대 그래프 인스턴스 생성 : 기본 값으로 선 그래프 생성
## - align : 막대와 라벨 위치 설정
## - width : 막대 너비 설정 [기] 0.8
## - color : 막대 색상 설정
## 색상코드 값
colors = [ "#ee00ff", "#00ff3c"]


## width(수직) <-> height(수평)
bar = plt.barh(sumSR.index, sumSR.values, align='edge', height=0.3,color=colors)


## ha : 수평 정렬 ()
## va : 수직 정렬
print(bar.patches)
for rect in bar : 
    # print(rect.get_height(), rect.get_width(), rect.get_x())
    h,w,x,y = rect.get_height(), rect.get_width(), rect.get_x(), rect.get_y()
    plt.text(x+w/2.0,y+h/2.0,f'{h}',fontdict={'fontsize' : 10, 'color' : 'blue'},
             ha='center', va='center')

## 그래프 기본요소
plt.title('컬럼별 합계')
plt.xlabel('컬럼')
plt.ylabel('합계')


## 화면에 그래프 출력
plt.show()
```

```python
## ---------------------------------------------------------------------
## [3] 응용 막대 그래프 - 여러개 데이터 비교
## ---------------------------------------------------------------------
## 데이터 준비
xData = pd.Series(range(1,8,3))     ## 1, 4, 7
yData1 = pd.Series([50,40,90])
yData2 = pd.Series([80,90,60])
yData3 = pd.Series([20,50,30])

# 시각화 준비
W,W2 = 0.8, 0.8*2

plt.bar(xData,  yData1, width=W)        # [1,    4,    7  ]
plt.bar(xData+W,  yData2, width=W)      # [1.8,  4.8,  7.8]
plt.bar(xData+W2,  yData3, width=W)     # [2.6,  5.6,  8.6]

plt.xticks(xData+W, ['1970', '1980','1990'])
plt.show()
```
