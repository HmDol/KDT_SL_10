<hr> 8. apply / lambda / 사용자 정의 함수 (apply_test) <hr>

```python
# === [8] apply / lambda / 사용자 정의 함수 ===
df = pd.read_csv('./EXAM_DATA/apply_test.csv')
display(df)
q8_1_sqrt = df['score1'].apply(lambda x : x ** 2)
# display(q8_1_sqrt)
q8_2_row_sum = df[['score1', 'score2']].apply(axis=1, func='sum')
# display(q8_2_row_sum)
q8_3_fullname = df[['name','age']].apply(axis=1, func=lambda x : f"{x['name']}_{x['age']}")
# display(q8_3_fullname)
q8_4_age_group = df['age'].apply(lambda x : '성인' if x >= 30 else '청년')
# display(q8_4_age_group)
def num_mean(df):
    return df.mean(numeric_only=True) ## select_dtypes 도 될듯
q8_5_num_mean = num_mean(df)
# display(q8_5_num_mean)
```
