### [ 인덱스 다루기 ]
- 기본 인덱스
    * .set_index()      <-- 컬럼을 인덱스로 설정>
    * .reindex()        <-- 인덱스 재구성(일부 변경, 결측치 발생 가능)>
    * .reset_index()    <-- 위치 인덱스(RangeIndex)로 변경. 기존 인덱스는 컬럼 추가>


[1] 모듈 로딩 및 데이터 준비 <hr>

```python
## 모듈 로딩
import pandas as pd
```

```python
## 데이터 준비
DATA_FILE = '../Data/학생관리부.xlsx'

## excel --> DF 로딩
## skiprows : 상단의 행 버리기
## index_col : 해당 컬럼 index로 설정
# dataDF = pd.read_excel(DATA_FILE, skiprows=2, index_col=0)
dataDF = pd.read_excel(DATA_FILE, skiprows=2)
```

```python
## 확인
display(dataDF.head())
```

[2] DF에서 인덱스관련 메서드 다루기 <hr>

```python
## ----------------------------------------------------------
## [2-1] 컬럼을 인덱스로 설정 : .set_index()
## ----------------------------------------------------------
dataDF2 = dataDF.set_index(['이름']).sort_index().copy()

dataDF2
```

```python
## -> [실습] 석차 컬럼이 인덱스 설정 및 내림차순 정렬
display(dataDF.index)

dataDF3 = dataDF.set_index([' 석차']).sort_index(ascending=False)
dataDF3
```

```python
## -> [매개변수 활용] 특정 컬럼이 행인덱스 가능한지 검사 진행
##                  대량의 데이터 경우 행 중복값
# dataDF3 = dataDF.set_index(['학년'], verify_integrity = True)
```

```python
## [문제] 2개 이상의 컬럼을 조합해서 행인덱스로 설정된 DF 설정
dataDF4 = dataDF.set_index(['이름', '학년'], verify_integrity=True)
dataDF4
```

```python
## ----------------------------------------------------------
## [2-2] 인덱스 재구성/재배치 : .reindex(index = [], columns = [])
##                           .reindex(labels = [], axis = 0/1)
## ----------------------------------------------------------
display(dataDF)
```

```python
## 이름, 학년, 반, 석차 ==> 학년 반 이름 석차 열이름/ 열인덱스를 재배치/재구성
# dataDF.reindex(labels=['학년', '반', '이름', ' 석차'], axis = 1)
dataDF.reindex(columns=['학년', '반', '이름', ' 석차'])
```

```python
## 이름, 학년, 반, 석차 ==> 학교, 이름, 학년, 반, 번호 열이름/ 열인덱스를 재배치/재구성
## 기존 없던 것 추가 : 학교, 번호
## 기존 있던 것 추가 : 석차 
testDF = dataDF.reindex(labels=['학교','이름','학년','반','번호'], axis='columns')
testDF['학교'] = ['경북대' for _ in range(len(dataDF.index))]
testDF
```

```python
## ----------------------------------------------------------
## [2-3] 인덱스 초기화 : .reset_index()
##                     기존 행인덱스는 컬럼으로 추가됨
## ----------------------------------------------------------
display(dataDF)
```

```python
## 인덱스 초기화 : [기본] 행인덱스가 => 컬럼으로 추가됨
dataDF2 = dataDF.reset_index()
dataDF2
```

```python
## 인덱스 초기화 : [설정] 행인덱스 => 컬럼추가 x, drop=True
dataDF2 = dataDF.reset_index(drop=True)
dataDF2
```

```python
## -------------------------------------------------
## [실습] 행인덱스 초기화
## -------------------------------------------------
## => 4행 2열
dataDF = pd.DataFrame([ ('bird', 389.0),
                        ('bird', 24.0),
                        ('mammal', 80.5),
                        ('mammal', pd.NA)],
                        index=['falcon', 'parrot', 'lion', 'monkey'],
                        columns=('class', 'max_speed'))
display(dataDF)
```

```python
## => [실습] 현재 행인덱스를 초기화 : 컬럼으로 전달되는 행인덱스 삭제 여부 결정
##                                필요 없는 데이터 drop=True
##                                필요 있는 데이터면 drop=False[기본]
dataDF.reset_index(names='kind')
```
