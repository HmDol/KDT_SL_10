### [ 수치형과 범주형  관계 ]
- 범주형(또는 텍스트형) 변수와 수치형 변수 간의 관계 => 통계적 측도(association measure)
- 수치형 변수와 수치형 변수 간의 관계 => 상관계수

```python
## ------------------------------------------------------------
## 1. 모듈 로딩 및 데이터 준비
## ------------------------------------------------------------
import pandas as pd
import numpy as np
import sys
sys.path.append(r'C:\Users\kdt008\Desktop\KDT_10SL\EX_DATA_VS\COMM')
from datautil import *

DATA_FILE = '../Data/titanic_train.csv'

## 데이터 로딩
dataDF = pd.read_csv(DATA_FILE)

## 본사본 준비
dataDF2 = dataDF.copy()
```

```python
## ------------------------------------------------------------
# 3) 타겟과 후보 범주형 컬럼들
## ------------------------------------------------------------
num_col = "Age"
cat_cols = ["Pclass", "Sex", "Embarked"]

rows = []
for c in cat_cols:
    eta2, ssb, sst, k = eta_squared_np(dataDF2, c, num_col)
    rows.append({
        "cat_col": c,
        "eta_squared": eta2,
        "SS_between": ssb,
        "SS_total": sst,
        "num_groups": k
    })

for r in rows:
    print(r['cat_col'], r['eta_squared'])
```
